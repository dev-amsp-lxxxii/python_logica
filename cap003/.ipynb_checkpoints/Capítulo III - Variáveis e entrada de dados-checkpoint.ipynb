### 3.2 - Variáveis numéricas

```python
print(1_000)
print(1_000_000)
print(1_0_0)
print(1_980.10)
```

```python
# 531
print(5 * 10**2 + 3 * 10**1 + 1 * 10**0)
print(5 * 100 + 3 * 10 + 1 * 1)
print(500 + 30 + 1)
```

```python
# 1010 -> 10
print(1 * 2**3 + 0 * 2**2 + 1 * 2**1 + 0 * 2**0)
print(1 * 8 + 0 * 4 + 1 * 2 + 0 * 1)
print(8 + 0 + 2 + 0)
```

```python
# Para precisão procurar sobre decimal e fractions
3 * 0.1
```

```python
a = 0b11 # base 2 - binário
b = 0x11 # base 16 - hexadecimal
c = 0o11 # base 8 - octal
print(f"a = 0b11 -> {a}")
print(f"b = 0x11 -> {b}")
print(f"c = 0o11 -> {c}")
```
