### 3.3 Variáveis do tipo Lógico

```python
resultado = True
aprovado = False
print(f"resultado: {resultado} e aprovado: {aprovado}")
```

```python
# Operadores relacionais

# == igualdade
# >  maior que
# <  menor que
# != diferente
# >= maior ou igual que
# <= menor ou igual que
```

```python
opr_a = 1
opr_b = 5
opr_c = 2
opr_d = 1

print(f"A é igual a B? {opr_a == opr_b}")
print(f"B é maior que A? {opr_b > opr_a}")
print(f"A é menor que B? {opr_a < opr_b}")
print(f"A é igual a D? {opr_a == opr_d}")
print(f"B é maior ou iguala a A? {opr_b >= opr_a}")
print(f"C é menor ou igual a B? {opr_c <= opr_b}")
print(f"D é diferente de A? {opr_d != opr_a}")
print(f"D é diferente de B? {opr_d != opr_b}")
```

```python
nota = 8.0
media = 7.0
aprovado = nota >= media
print(f"Aprovado: {aprovado}")
```

```python
# Exercício 3.2
"""
Complete a tabela a seguir, respondendo True ou False.
Considere  ex_a = 4, ex_b = 10, ex_c = 5.0, ex_d = 1 e ex_f = 5 
"""

ex_a = 4; ex_b = 10; ex_c = 5.0; ex_d = 1; ex_f = 5

print(f"a == c -> {ex_a} == {ex_c}: {ex_a == ex_c}")
print(f"a < b -> {ex_a} < {ex_b}: {ex_a < ex_b}")
print(f"d > b -> {ex_d} > {ex_b}: {ex_d > ex_b}")
print(f"c != f ->  {ex_c} != {ex_f}: {ex_c != ex_f}")
print(f"a == b -> {ex_a} == {ex_b}: {ex_a == ex_b}")
print(f"c < d -> {ex_c} < {ex_d}: {ex_c < ex_d}")
print(f"b > a -> {ex_b} > {ex_a}: {ex_b > ex_a}")
print(f"c >= f -> {ex_c} >= {ex_f}: {ex_c >= ex_f}")
print(f"f >= c -> {ex_f} >= {ex_c}: {ex_c >= ex_f}")
print(f"c <= c -> {ex_c} <= {ex_c}: {ex_c <= ex_f}")
print(f"c <= f -> {ex_c} <= {ex_f}: {ex_c <= ex_c}")
```

```python
# Exercício 3.6
"""
Escreva uma expressão que será utilizado para decidir se um aluno foi ou não aprovado.
Para ser aprovado, todas as médias do aluno devem ser maiores ou iguais a 7. Considere que o aluno
cursa apenas três matérias, e que a nota de cada uma está armazenada nas seguintes variáveis:
matéria01, matéria02, matéria03.
"""

nome = "Adelmo Menezes"

materia01 = 6.0;  materia02 = 7.5; materia03 = 9.5

resultado = (materia01 + materia02 + materia03) / 3

print(f"O aluno {nome} está aprovado? {resultado >= 7.0} com média: {resultado:.2f}")
```

```python
# 3.4.1.2 - Composição
x_idade = 15
imprimir_idade = "João tem %d anos." % x_idade
print(imprimir_idade)

# principais marcadores -> %d inteiros - %s string - %f decimais
```

```python
idade = 43
print("[%d]" % idade)
print("[%03d]" % idade)
print("[%3d]" % idade)
print("[%-3d]" % idade)
```

```python
print("R$%f" % 7)
print("R$%5f" % 7)
print("R$%5.2f" % 7)
print("R$%10.5f" % 7)
```

```python
composicao_nome = "Adelmo"
composicao_idade = 43
composicao_saldo = 51.34
```

```python
# Marcador
print("%s tem %d anos e R$%f no bolso." % (composicao_nome, composicao_idade, composicao_saldo))
print("[%12s] tem [%3d] anos e [R$%5.2f] no bolso." % (composicao_nome, composicao_idade, composicao_saldo))
print("[%12s] tem [%03d] anos e [R$%5.2f] no bolso." % (composicao_nome, composicao_idade, composicao_saldo))
print("[%-12s] tem [%-3d] anos e [R$%-5.2f] no bolso." % (composicao_nome, composicao_idade, composicao_saldo))
```

```python
# format()
print("{} tem {} anos e R${} no bolso.".format(composicao_nome, composicao_idade, composicao_saldo))
print("[{:12}] tem [{:3}] anos e [R${:5.2f}] no bolso.".format(composicao_nome, composicao_idade, composicao_saldo))
print("[{:12}] tem [{:03}] anos e [R${:5.2f}] no bolso.".format(composicao_nome, composicao_idade, composicao_saldo))
print("[{:<12s}] tem [{:<3}] anos e [R${:5.2f}] no bolso.".format(composicao_nome, composicao_idade, composicao_saldo))
```

```python
print("{1} - {0} - {2}".format("A", "B", "C"))
print("{2} - {2} - {0} - {1}".format("A", "B", "C"))
```

```python
# f-string
print(f"{composicao_nome} tem {composicao_idade} anos e R${composicao_saldo} no bolso.")
print(f"[{composicao_nome:12}] tem [{composicao_idade:3}] anos e [R${composicao_saldo:5.2f}] no bolso.")
print(f"[{composicao_nome:12}] tem [{composicao_idade:03}] anos e [R${composicao_saldo:5.2f}] no bolso.")
print(f"[{composicao_nome:<12}] tem [{composicao_idade:<3}] anos e [R${composicao_saldo:5.2f}] no bolso.")
```

```python
# fatiamento
fatia_s = "ABCDEFGHI"
print(fatia_s[0:2])
print(fatia_s[1:2])
print(fatia_s[2:4])
print(fatia_s[0:5])
print(fatia_s[1:8])
```
